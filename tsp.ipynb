```python
#Brute force approach
# Runtime: O(n!)
from itertools import permutations
from matplotlib.lines import VertexSelector
import numpy as np
import itertools  
from timeit import default_timer as timer
matrixSize = 10
step = 0
times = 0 

#matrixSize = 13

dist = np.array([
    [0, 20, 42, 25, 10, 20, 11, 10, 100, 1],
    [20, 0, 30, 34, 15, 40, 22, 20, 90, 4],
    [42, 30, 0, 10, 20, 50, 33, 30, 80, 9],
    [25, 34, 10, 0, 30, 60, 44, 40, 70, 16],
    [10, 15, 20, 30, 0, 70, 55, 50, 60, 25],
    [20, 40, 50, 60, 70, 0, 6, 60, 50, 36],
    [11, 22, 33, 44, 55, 6, 0, 70, 40, 49],
    [10, 20, 30, 40, 50, 60, 70, 0, 30, 64],
    [100, 90, 80, 70, 60, 50, 40, 30, 0, 81],
    [ 1, 4, 9, 16, 25, 36, 49, 64, 81, 0]
])

n = len(dist)

INT_MAX = 9999999

def bruteforcetsp(start):
 
  vertex = []
  for i in range(n):
    if i != start:
      vertex.append(i)

  minpath = INT_MAX
  nextperm = itertools.permutations(vertex)
  for i in nextperm:
    currentLength = 0
    k = start #sets the start as k
    for j in i: # add up distance in permutation from start to next 
      currentLength += dist[k][j]
      k = j # after adding start to next it sets next to k and goes back in for loop
  #add final length back to start position 
    currentLength += dist[k][start]

    if currentLength < minpath:
      pathTaken = i
    minpath = min(minpath,currentLength)
    


  return minpath, pathTaken

start = timer()
print(bruteforcetsp(0))
end = timer()
print(str(end - start) + "seconds")
```

```python
#Brute Force with top-down and bitmask approach
#Runtime: o(n!)
import numpy as np
import time
start_time = time.time()
from typing_extensions import runtime

#distance matrix defining the distances between cities in the graph
dist = np.array([
    [0, 20, 42, 25, 10, 20, 11, 10, 100, 1],
    [20, 0, 30, 34, 15, 40, 22, 20, 90, 4],
    [42, 30, 0, 10, 20, 50, 33, 30, 80, 9],
    [25, 34, 10, 0, 30, 60, 44, 40, 70, 16],
    [10, 15, 20, 30, 0, 70, 55, 50, 60, 25],
    [20, 40, 50, 60, 70, 0, 6, 60, 50, 36],
    [11, 22, 33, 44, 55, 6, 0, 70, 40, 49],
    [10, 20, 30, 40, 50, 60, 70, 0, 30, 64],
    [100, 90, 80, 70, 60, 50, 40, 30, 0, 81],
    [ 1, 4, 9, 16, 25, 36, 49, 64, 81, 0]
])

n = len(dist)
dist_size = 2**n

max_val = 999999


# Checking if all cities have been visited
visited_all = (1<<n) -1

#creating a function to compare bitmask with 1s representing visited and
# 0s representing not visited to bitmask full of 1s to create a stopping criteria
def tsp(mask, pos):
  if (mask == visited_all):
    return dist[pos][0]
   
  ans = max_val
  # Go to city NOT visited previously
  for city in range(n):
    if ((mask & (1<<city))==0):
      newAns = dist[pos][city] + tsp(mask|(1<<city), city)
      ans = min(ans, newAns)
  return ans

print("The minimum distance to travel to " + str(n) + " cities is " + str(tsp(1,0)))
print("%s seconds" % (time.time() - start_time))
```

```python
#Dynamic Programing Top-Down
# Runtime: o(n^2)
import numpy as np
import time
start_time = time.time()
from typing_extensions import runtime

INT_MAX = 999999
# adjacency matrix defining the distances between cities in the graph
dist = np.array([
    [0, 20, 42, 25, 10, 20, 11, 10, 100, 1],
    [20, 0, 30, 34, 15, 40, 22, 20, 90, 4],
    [42, 30, 0, 10, 20, 50, 33, 30, 80, 9],
    [25, 34, 10, 0, 30, 60, 44, 40, 70, 16],
    [10, 15, 20, 30, 0, 70, 55, 50, 60, 25],
    [20, 40, 50, 60, 70, 0, 6, 60, 50, 36],
    [11, 22, 33, 44, 55, 6, 0, 70, 40, 49],
    [10, 20, 30, 40, 50, 60, 70, 0, 30, 64],
    [100, 90, 80, 70, 60, 50, 40, 30, 0, 81],
    [ 1, 4, 9, 16, 25, 36, 49, 64, 81, 0]
])

n = len(dist)
m = len(dist[0])
dist_size = 2 ** n
dp = np.zeros((dist_size, m))
tour = []

# print(dp)

# Checking if all cities have been visited
visited_all = (1 << n) - 1


def tsp(mask, pos):
    if (mask == visited_all):
      return dist[pos][0]

    # Lookup if route has been traversed
    if dp[mask][pos] != -1:
        return dp[mask][pos]

    ans = INT_MAX
    # Go to city NOT visited previously
    for city in range(n):
        if ((mask & (1 << city)) == 0):
            newAns = dist[pos][city] + tsp(mask | (1 << city), city)
            ans = min(ans, newAns)         

    dp[mask][pos] = ans
    return ans

for city in range((1 << n)):
    for j in range(n):
        dp[city][j] = -1


print("The minimum distance to travel to " + str(n) + " cities is " + str(tsp(1, 0)))
print("%s seconds" % (time.time() - start_time))
```
